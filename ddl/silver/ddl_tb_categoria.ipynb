```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_categoria"

full_table_name_destination = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {full_table_name_destination} (
    cd_categoria BIGINT COMMENT 'Identificador da categoria.',
    ds_categoria STRING COMMENT 'Descrição da categoria.',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingestão do registro na tabela.',
    dc_origem STRING COMMENT 'Identificação do sistema ou arquivo fonte dos dados.',
    CONSTRAINT pk_categoria PRIMARY KEY (cd_categoria)
)
COMMENT 'Tabela silver de cadastro de categorias';
"""

spark.sql(query)
print(f"✅ Tabela {full_table_name_destination} criada com sucesso!")
```
