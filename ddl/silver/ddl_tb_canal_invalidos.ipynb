```python
CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_canal_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_canal BIGINT COMMENT 'Identificador gerado (PK).',
    nm_canal VARCHAR(50) COMMENT 'Nome do canal de comunicação.',
    st_ativo BOOLEAN COMMENT 'Status do canal (Ativo/Inativo).',
    st_ativo_raw STRING COMMENT 'Valor original do status (para debug).',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingestão do registro.',
    dc_origem STRING COMMENT 'Identificação do sistema ou arquivo fonte dos dados.',
    flag_nome_valido BOOLEAN COMMENT 'Indica se o nome do canal é válido.',
    flag_status_valido BOOLEAN COMMENT 'Indica se o status foi mapeado corretamente.',
    flag_qualidade STRING COMMENT 'Status de qualidade: OK ou ERRO.',
    CONSTRAINT pk_canal_invalidos PRIMARY KEY (cd_canal)
)
COMMENT 'Tabela silver de auditoria - registros de canais rejeitados por problemas de qualidade';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
