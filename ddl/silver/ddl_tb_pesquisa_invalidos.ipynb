```python
CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_pesquisa_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_pesquisa BIGINT COMMENT 'Identificador da pesquisa (PK).',
    cd_chamado BIGINT COMMENT 'Vínculo com o chamado avaliado.',
    nu_nota SMALLINT COMMENT 'Nota atribuída (1 a 5, pode ser NULL).',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingestão do registro.',
    dc_origem STRING COMMENT 'Identificação do sistema ou arquivo fonte dos dados.',
    flag_pk_valida BOOLEAN COMMENT 'Indica se o ID da pesquisa é válido.',
    flag_fk_valida BOOLEAN COMMENT 'Indica se o ID do chamado é válido.',
    flag_nota_valida BOOLEAN COMMENT 'Indica se a nota está no range 1-5 ou é NULL.',
    flag_qualidade STRING COMMENT 'Status de qualidade: OK ou ERRO.',
    CONSTRAINT pk_pesquisa_invalidos PRIMARY KEY (cd_pesquisa)
)
COMMENT 'Tabela silver de auditoria - registros de pesquisas rejeitados por problemas de qualidade';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
