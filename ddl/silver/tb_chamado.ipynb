```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_chamado"

full_table_name_destination = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {full_table_name_destination} (
    cd_chamado BIGINT COMMENT 'Identificador do chamado',
    cd_cliente BIGINT COMMENT 'Referência ao cliente',
    ds_motivo_rel STRING COMMENT 'Motivo relatado',
    nm_canal_origem STRING COMMENT 'Nome do canal de entrada',
    st_resolvido BOOLEAN COMMENT 'Indicador de resolução',
    dh_abertura TIMESTAMP COMMENT 'Data e hora de abertura',
    dh_inicio TIMESTAMP COMMENT 'Data e hora do início real',
    dh_fim TIMESTAMP COMMENT 'Data e hora de conclusão',
    tm_espera STRING COMMENT 'Tempo em fila até atendimento',
    tm_duracao STRING COMMENT 'Duração total do atendimento',
    cd_atendente BIGINT COMMENT 'Referência ao atendente',
    ingestion_timestamp TIMESTAMP,
    origem STRING,
    CONSTRAINT pk_chamados PRIMARY KEY (cd_chamado)
)
COMMENT 'Tabela silver de histórico de chamados';
"""

spark.sql(query)
print(f"✅ Tabela {full_table_name_destination} criada com sucesso!")
```
