```python
%python
CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_motivo_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
%python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_motivo BIGINT COMMENT 'Identificador √∫nico do motivo.',
    ds_motivo STRING COMMENT 'Descri√ß√£o do motivo.',
    ds_categoria STRING COMMENT 'Categoria do motivo (CADASTRAL, CARTAO, FINANCEIRO, etc).',
    ds_criticidade STRING COMMENT 'N√≠vel de criticidade (BAIXA, MEDIA, ALTA).',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingest√£o do registro.',
    dc_origem STRING COMMENT 'Identifica√ß√£o do sistema ou arquivo fonte dos dados.',
    flag_qualidade STRING COMMENT 'Status de qualidade: OK ou ERRO.',
    motivo_rejeicao STRING COMMENT 'Descri√ß√£o dos motivos de rejei√ß√£o do registro.',
    CONSTRAINT pk_motivo_invalidos PRIMARY KEY (cd_motivo)
)
COMMENT 'Tabela silver de auditoria - registros de motivos rejeitados por problemas de qualidade';
"""

spark.sql(query)
print(f"‚úÖ Tabela {nome_tabela_destino} criada com sucesso!")
```
