```sql
%python
CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_motivo_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

### DDL - Tabela de Auditoria (Motivos Invalidos)
Cria tabela de auditoria para motivos rejeitados:
- **motivo_rejeicao**: Descricao textual dos problemas encontrados
- **flag_qualidade**: Sempre "ERRO"

**Importante**: Rastreia motivos com categorias ou criticidades invalidas.

### Configuracao
Define catalogo, schema e nome da tabela de auditoria de motivos invalidos.

```sql
%python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_motivo BIGINT COMMENT 'Identificador único do motivo.',
    ds_motivo STRING COMMENT 'Descrição do motivo.',
    ds_categoria STRING COMMENT 'Categoria do motivo (CADASTRAL, CARTAO, FINANCEIRO, etc).',
    ds_criticidade STRING COMMENT 'Nível de criticidade (BAIXA, MEDIA, ALTA).',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingestão do registro.',
    dc_origem STRING COMMENT 'Identificação do sistema ou arquivo fonte dos dados.',
    flag_qualidade STRING COMMENT 'Status de qualidade: OK ou ERRO.',
    motivo_rejeicao STRING COMMENT 'Descrição dos motivos de rejeição do registro.',
    CONSTRAINT pk_motivo_invalidos PRIMARY KEY (cd_motivo)
)
COMMENT 'Tabela silver de auditoria - registros de motivos rejeitados por problemas de qualidade';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
