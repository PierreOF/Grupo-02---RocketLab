```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_chamado_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_chamado BIGINT COMMENT 'Identificador do chamado (PK).',
    cd_cliente STRING COMMENT 'Hash SHA256 do cliente.',
    cd_motivo BIGINT COMMENT 'Identificador do motivo.',
    cd_canal BIGINT COMMENT 'Identificador do canal.',
    st_resolvido BOOLEAN COMMENT 'Indicador de resolução.',
    tm_espera BIGINT COMMENT 'Tempo de espera em segundos.',
    tm_duracao BIGINT COMMENT 'Duração do atendimento em segundos.',
    cd_atendente BIGINT COMMENT 'Identificador do atendente (NULL se automatizado).',
    dt_ingestion TIMESTAMP COMMENT 'Data e hora da ingestão.',
    dc_origem STRING COMMENT 'Identificação da origem.',
    flag_pk_valida BOOLEAN COMMENT 'Flag: cd_chamado é válido.',
    flag_fk_cliente_valida BOOLEAN COMMENT 'Flag: cd_cliente é válido.',
    flag_fk_motivo_valida BOOLEAN COMMENT 'Flag: cd_motivo foi encontrado.',
    flag_fk_canal_valida BOOLEAN COMMENT 'Flag: cd_canal foi encontrado.',
    flag_qualidade STRING COMMENT 'Status: OK ou ERRO.',
    CONSTRAINT pk_chamado_invalidos PRIMARY KEY (cd_chamado)
)
COMMENT 'Tabela silver de auditoria - chamados rejeitados';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
