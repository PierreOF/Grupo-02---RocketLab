```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_atendente"

full_table_name_destination = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {full_table_name_destination} (
    cd_atendente BIGINT COMMENT 'Identificador único do atendente.',
    nm_atendente STRING COMMENT 'Nome completo do funcionário.',
    nu_nivel SMALLINT COMMENT 'Nível de senioridade do atendimento.',
    dt_ingestion TIMESTAMP,
    dc_origem STRING,
    CONSTRAINT pk_atendente PRIMARY KEY (cd_atendente)
)
COMMENT 'Tabela silver de cadastro de atendentes';
"""

spark.sql(query)
print(f"✅ Tabela {full_table_name_destination} criada com sucesso!")
```
