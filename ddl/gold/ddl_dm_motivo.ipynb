```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "gold"
TABELA_DESTINO = "dm_motivo"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python


query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_motivo BIGINT COMMENT 'Identificador único do motivo. (PK)',
    ds_motivo VARCHAR(100) COMMENT 'Descrição resumida do motivo.',
    ds_categoria VARCHAR(50) COMMENT 'Categoria macro do motivo do chamado.',
    ds_criticidade VARCHAR(50) COMMENT 'Identificador único da criticidade.',

    -- Definição de Chave (Constraint)
    CONSTRAINT pk_motivo PRIMARY KEY (cd_motivo)
)
USING DELTA
COMMENT 'Tabela de dimensão de motivos de chamado/ocorrência.';
"""


spark.sql(query)

print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
