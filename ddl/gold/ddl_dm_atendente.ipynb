```python
from pyspark.sql import functions as F

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "dm_atendente"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

### DDL - Dimensao Atendente
Cria dimensao de funcionarios para Star Schema:
- **cd_atendente** (PK): Codigo do funcionario
- **ds_perfil**: Coluna derivada que classifica perfil ("Generalista (N1)" ou "Especialista (N2)")

**Importante**: Dimensao usada nos fatos ft_atendimento_geral e ft_performance_agente.

### Configuracao
Define catalogo, schema e nome da dimensao de atendentes (camada Gold - Star Schema).

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    cd_atendente BIGINT COMMENT 'PK: Código do funcionário.',
    nm_atendente STRING COMMENT 'Nome do funcionário.',
    nu_nivel SMALLINT COMMENT 'Nível (1 ou 2).',
    ds_perfil STRING COMMENT 'Perfil descritivo (Especialista N2 ou Especialista N1).'
)
COMMENT 'Dimensão de cadastro da equipe humana.';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
