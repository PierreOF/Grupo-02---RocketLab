```python
from pyspark.sql import SparkSession

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "dm_chamado"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

### DDL - Dimensao Chamado
Cria dimensao de protocolos de chamados:
- **cd_chamado** (PK): Codigo original do chamado (numero do protocolo)
- **tm_espera**: Tempo em fila ate atendimento (segundos)
- **tm_duracao**: Duracao total do atendimento (segundos)

**Importante**: Dimensao degenerada (degenerate dimension) - armazena atributos do evento transacional.

### Configuracao
Define catalogo, schema e nome da dimensao de chamados (camada Gold - Star Schema).

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    cd_chamado BIGINT COMMENT 'PK: Código original do chamado (Protocolo).',
    
    tm_espera BIGINT COMMENT 'Tempo em fila até atendimento.',
    tm_duracao BIGINT COMMENT 'Duração total do atendimento.'
)
COMMENT 'Dimensão de cadastro dos chamados (Protocolos e Datas).';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
