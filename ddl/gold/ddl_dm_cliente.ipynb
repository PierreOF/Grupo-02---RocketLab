```python
from pyspark.sql import functions as F

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "dm_cliente"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    cd_cliente STRING COMMENT 'PK: Hash único do cliente (36 chars).',
    nm_cliente VARCHAR(150) COMMENT 'Nome completo do cliente.',
    ds_email VARCHAR(150) COMMENT 'E-mail de contato.',
    nm_regiao VARCHAR(50) COMMENT 'Região geográfica (Ex: Sudeste, Norte).',
    nu_idade SMALLINT COMMENT 'Idade do cliente.',
    ds_faixa_etaria STRING COMMENT 'Classificação etária (Geração Z, Senior, etc).'
)
COMMENT 'Dimensão de perfil demográfico dos clientes.';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
