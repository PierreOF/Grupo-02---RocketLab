```python
from pyspark.sql import functions as F

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "dm_canal"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

### DDL - Dimensao Canal
Cria dimensao de canais de atendimento:
- **cd_canal** (PK): Codigo do canal
- **ds_tipo_interacao**: Coluna derivada que classifica canais em:
  - "Automatizado" (Chatbot, URA)
  - "Humanizado" (Atendimento Inicial N1, Atendimento Especializado N2)

**Importante**: Permite comparar performance Bot vs Humano (tempo de espera, CSAT, custo).

### Configuracao
Define catalogo, schema e nome da dimensao de canais (camada Gold - Star Schema).

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    cd_canal BIGINT COMMENT 'PK: Código do canal.',
    nm_canal VARCHAR(50) COMMENT 'Nome do canal (URA, Whatsapp, App).',
    st_ativo BOOLEAN COMMENT 'Status ativo/inativo.',
    ds_tipo_interacao STRING COMMENT 'Classificação: Automatizado ou Humanizado.'
)
COMMENT 'Dimensão de canais de atendimento.';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
