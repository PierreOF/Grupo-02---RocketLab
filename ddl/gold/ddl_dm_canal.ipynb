```python
from pyspark.sql import functions as F

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "gold"
TABELA_DESTINO = "dm_canal"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_canal BIGINT COMMENT 'Identificador gerado (sugerido).',
    nm_canal VARCHAR(50) COMMENT 'Nome do canal de comunicação.',
    st_ativo BOOLEAN COMMENT 'Status do canal (Ativo/Inativo).',
    ds_tipo_interacao STRING COMMENT 'Classificação macro: Automatizado vs Humanizado.',
    dt_ingestion TIMESTAMP,
    dc_origem STRING,
    CONSTRAINT pk_dm_canal PRIMARY KEY (cd_canal)
)
COMMENT 'Tabela dimensão de canais de atendimento (Gold)';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
