```python
from pyspark.sql import functions as F

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "dm_canal"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    cd_canal BIGINT COMMENT 'PK: Código do canal.',
    nm_canal STRING COMMENT 'Nome do canal (URA, Whatsapp, App).',
    st_ativo BOOLEAN COMMENT 'Status ativo/inativo.',
    ds_tipo_interacao STRING COMMENT 'Classificação: Automatizado ou Humanizado.'
)
COMMENT 'Dimensão de canais de atendimento.';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
