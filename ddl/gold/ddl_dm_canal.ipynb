```python
from pyspark.sql import functions as F
from pyspark.sql.types import TimestampType

CATALOGO_DESTINO = "v_credit"
SCHEMA_DESTINO = "gold"
TABELA_DESTINO = "dm_canal"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    cd_canal BIGINT COMMENT 'Identificador gerado (sugerido). (PK)',
    nm_canal VARCHAR(50) COMMENT 'Nome do canal de comunicação.',
    st_ativo BOOLEAN COMMENT 'Status do canal (Ativo/Inativo).',

    -- Definição de Chave (Constraint)
    CONSTRAINT pk_canal PRIMARY KEY (cd_canal)
)
USING DELTA
COMMENT 'Tabela de dimensão de canais de comunicação.';
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
