```python
from pyspark.sql import functions as F

CATALOGO = "v_credit"
SCHEMA = "gold"
TABELA = "ft_atendimento_geral"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{TABELA}"
```

```python
spark.sql(f"""
CREATE OR REPLACE TABLE {FULL_NAME} (
    pk_fato_atendimento STRING COMMENT 'PK: Hash único da linha.',
    cd_cliente STRING COMMENT 'FK: Liga com dm_cliente.',
    cd_motivo BIGINT COMMENT 'FK: Liga com dm_motivo.',
    cd_canal BIGINT COMMENT 'FK: Liga com dm_canal.',
    cd_atendente BIGINT COMMENT 'FK: Liga com dm_atendente.',
    
    cd_chamado BIGINT COMMENT 'Degenerate Dimension: Código original.',
    
    dt_referencia DATE COMMENT 'Data do fato.',
    nu_hora_dia SMALLINT COMMENT 'Hora (0-23).',
    
    val_tempo_espera BIGINT COMMENT 'Métrica: Espera (seg).',
    val_tempo_atendimento BIGINT COMMENT 'Métrica: Duração (seg).',
    val_custo DECIMAL(12,10) COMMENT 'Métrica: Custo unitário preciso.',
    val_nota_csat SMALLINT COMMENT 'Métrica: Nota (1-5).',
    
    fl_experiencia_negativa SMALLINT COMMENT 'Flag (0/1): Espera > 5min ou Nota <= 2.',
    st_resolvido SMALLINT COMMENT 'Flag (0/1): Problema resolvido.',
    
    dt_ingestion TIMESTAMP COMMENT 'Data de carga.',
    dc_origem STRING COMMENT 'Identificação do sistema ou arquivo fonte dos dados.'
)
COMMENT 'Fato principal de log transacional de atendimento.';
""")

print(f"✅ Tabela {FULL_NAME} criada com sucesso!")
```
